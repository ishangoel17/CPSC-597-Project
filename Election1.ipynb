```python
df['Score'].replace([np.inf, -np.inf], np.nan, inplace=True)

# Visualize sentiment scores
plt.figure(figsize=(10, 6))
sns.histplot(df['Score'].dropna(), kde=True, color='skyblue')
plt.title('Distribution of Sentiment Scores in the Dataset')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')

# Set x-axis and y-axis limits
plt.xlim(-30, 30)  # Set x-axis limits from -10 to 10
plt.ylim(0, 5)     # Set y-axis limits from 0 to 1

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='PartyName', y='Score', data=df, palette='muted')
plt.title('Sentiment Scores Distribution by Party Labels')
plt.xlabel('Party Labels')
plt.ylabel('Sentiment Score')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  # Increase figure size for better readability
sns.countplot(x='Language', data=df, palette='viridis')
plt.title('Language Distribution in the Dataset')
plt.xlabel('Language')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
